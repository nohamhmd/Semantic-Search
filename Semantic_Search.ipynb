# Preprocessing

```python
#df = df.drop(columns=['claps', 'reading_time', 'link'])
```

```python
def clean_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'http\S+', '', text)
    text = re.sub(r'<[^>]*>', '', text)
    text = re.sub(r'\s+', ' ', text)  # Remove extra whitespaces
    text = re.sub(r'[^\w\s]', '', text)  # Remove special characters (punctuation)
    text = text.replace('’', '')
    text = text.replace('‘', '')
    text = text.strip()  # Remove leading and trailing spaces

    return text
```

```python
# Remove stop words
stop_words = set(stopwords.words('english'))
def remove_stopwords(text):
    words = text.split()
    return ' '.join([word for word in words if word not in stop_words])
```

```python
# Apply Cleaning
df['text'] = df['text'].apply(clean_text)
df['title'] = df['title'].apply(clean_text)
df['text'] = df['text'].apply(remove_stopwords)
df['title'] = df['title'].apply(remove_stopwords)
```

```python
# Combine all text data into one string
text_data = ' '.join(df['text'])

# Create the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text_data)

# Display the word cloud using matplotlib
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Turn off axis
plt.show()
```

```python
# Combine all text data into one string
text_data = df['text']

# Initialize the vectorizer for unigrams
unigram_vectorizer = CountVectorizer(ngram_range=(1, 1), stop_words='english', max_features=10)
unigram_counts = unigram_vectorizer.fit_transform(text_data)

# Get top 10 unigrams
unigram_freq = pd.DataFrame(unigram_counts.toarray(), columns=unigram_vectorizer.get_feature_names_out())
unigram_sum = unigram_freq.sum().sort_values(ascending=False).head(10)

# Initialize the vectorizer for bigrams
bigram_vectorizer = CountVectorizer(ngram_range=(2, 2), stop_words='english', max_features=10)
bigram_counts = bigram_vectorizer.fit_transform(text_data)

# Get top 10 bigrams
bigram_freq = pd.DataFrame(bigram_counts.toarray(), columns=bigram_vectorizer.get_feature_names_out())
bigram_sum = bigram_freq.sum().sort_values(ascending=False).head(10)
```

```python
# Plotting top 10 unigrams
plt.figure(figsize=(10, 6))
unigram_sum.plot(kind='bar', color='lightblue')
plt.title('Top 10 Unigrams')
plt.xlabel('Unigrams')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plotting top 10 bigrams
plt.figure(figsize=(10, 6))
bigram_sum.plot(kind='bar', color='lightcoral')
plt.title('Top 10 Bigrams')
plt.xlabel('Bigrams')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
# Merge both column as it's convenient with our case, and offers more contextual meaning to be captured.
df['Content'] = df['title'] + ' ' + df['text']
```

```python
# Initialize the text splitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=512,  chunk_overlap=50)
```

```python
# Combine all text into a single string
all_text = " ".join(df['Content'].tolist())

# Split the combined text into chunks
text_chunks = text_splitter.split_text(all_text)
```

# Creating the vector database

```python
dir = 'VectorDB' # loacl
# Initialize SentenceTransformer model
embedding = HuggingFaceEmbeddings(model_name='all-MiniLM-L6-v2')

# Wrap the text chunks into Document objects
documents = [Document(page_content=text) for text in text_chunks]

# Initialize Chroma and load the documents with embeddings
vectordb = Chroma.from_documents(documents=documents,
                                 embedding=embedding,
                                 persist_directory=dir)
```

```python
# persiste the db to disk
vectordb.persist()
vectordb = None
```

```python
# Now we can load the persisted database from disk, and use it.
vectordb = Chroma(persist_directory=dir,
                  embedding_function=embedding)
```

# Semantic Search

```python
retriever = vectordb.as_retriever() # We can control the number of rerieved articles using this parameter --> search_kwargs={"k": 2}
```

```python
docs = retriever.get_relevant_documents("Machine Learning For Beginners")
```

# Hot KeyWords from Retrieved Articles

```python
for doc in docs:
    text = doc.page_content  # Get the content of the document
    keyphrases = kb.extract_keywords(text, keyphrase_ngram_range=(1, 3), highlight=True)
    print(keyphrases)
```
